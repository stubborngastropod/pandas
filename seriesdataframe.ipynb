```python
s[['나', '라']]
```

```python
np.random.seed(20)
sample2 = pd.Series(np.random.randint(100, 200, size=(15,)))
sample2
```

```python
sample2[sample2 <= 160]
```

```python
sample2[(sample2 >= 130) & (sample2 <= 170)]
```

```python
s = pd.Series(['apple', np.nan, 'banana', 'kiwi', 'gubong'], index=list('가나다라마'))
s
```

### Dataframe

```python
pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
```

```python
pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]], columns=['가', '나', '다'])
```

```python
data = {
    'name': ['Kim', 'Lee', 'Park'],
    'age': [24, 27, 34],
    'children': [2, 1, 3]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df.index)
print('---')
print(df.columns)
print('---')
print(df.values)
print('---')
print(df.dtypes)
print('---')
print(df.T)
```

```python
df.index = list('abc')
df
```

```python
df['name']
```

```python
df[['name', 'children']]
```

```python
df.rename(columns={'name': '이름'})
```

```python
df.rename(columns={'age': '나이'}, inplace=True)
df
```

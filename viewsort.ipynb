```python
df = sns.load_dataset("titanic")
df.head()
```

```python
cond1 = (df.age >= 30)
cond2 = (df.sex == 'male')
df.loc[cond1 & cond2].sort_values(by='fare', ascending=False).head(10)
```

```python
cond1 = (df.age >= 20) & (df.age < 40)
cond2 = (df['pclass'] < 3)
df.loc[cond1 & cond2, ['survived', 'pclass', 'age', 'fare']].head(10)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
cond1 = (tips['day'].isin(['Fri', 'Sat']))
cond2 = (tips['tip'] < 10)
tips.loc[cond1 & cond2, ['total_bill', 'tip', 'smoker', 'time']].head(10)
```
